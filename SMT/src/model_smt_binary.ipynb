```python
def at_least_one(bool_vars):
    return Or(bool_vars)

def at_most_one(bool_vars, name):
    constraints = []
    n = len(bool_vars)
    s = [Bool(f"s_{name}_{i}") for i in range(n - 1)]
    constraints.append(Or(Not(bool_vars[0]), s[0]))
    constraints.append(Or(Not(bool_vars[n-1]), Not(s[n-2])))
    for i in range(1, n - 1):
        constraints.append(Or(Not(bool_vars[i]), s[i]))
        constraints.append(Or(Not(bool_vars[i]), Not(s[i-1])))
        constraints.append(Or(Not(s[i-1]), s[i]))
    return And(constraints)

def exactly_one(bool_vars, name):
    return And(at_least_one(bool_vars), at_most_one(bool_vars, name))

```

```python
class Courier:
    def __init__(self, courier, items,low_bound,up_bound,d_low_bound, courier_size, item_size, distances):
        self.courier = courier
        self.items = items
        self.low_bound = low_bound
        self.up_bound = up_bound
        self.d_low_bound = d_low_bound
        self.courier_size = courier_size
        self.item_size = item_size
        self.distances = distances


instance = Courier(courier,items,low_bound, up_bound, d_low_bound, courier_size, item_size, distances)
```

```python
def mtps(instance):

      start_time = t.time()

      courier = instance.courier
      items = instance.items
      low_bound = instance.low_bound
      up_bound = instance.up_bound
      d_low_bound = instance.d_low_bound
      courier_size = instance.courier_size
      item_size = instance.item_size
      distances = instance.distances

      solver = Solver()

      couriers_loads = [Int(f'loads{i}') for i in range(courier)]
      m_dist = [Int(f"dist{i}") for i in range(courier)]
      x = [Array(f"asg{i}", IntSort(), IntSort()) for i in range(courier)]

      maximum = Int(f"max")

      # define the domain of x
      for k in range(courier):
            for i in range(items + 1):
                solver.add(x[k][i] >= 1)
                solver.add(x[k][i] <= items+1)

      # define the domain of maximum
      solver.add(maximum >= low_bound)
      solver.add(maximum <= up_bound)

      # define the domain for m_dist
      for k in range(courier):
            solver.add(m_dist[k] >= d_low_bound)
            solver.add(m_dist[k] <= up_bound)

      # maximum always grater than m_dist
      for k in range(courier):
            solver.add(maximum >= m_dist[k])


      # dopo lo zero son tutti zeri
      for k in range(courier):
             for i in range(items+1):
                   solver.add(Implies(x[k][i] == items+1, And([x[k][j] == items+1 for j in range(i+1, items+1)])))


      # Each item must be visited exactly once
      for j in range(1,items + 1):
            solver.add(exactly_one([x[k][i] == j for k in range(courier) for i in range(items + 1)], f"A{j}"))

      # Bin packing capacity
      #for k in range(courier):
            #solver.add(couriers_loads[k] == Sum([If(x[k][i] != items+1, item_size[x[k][i]], 0) for i in range(items+1)]))
      for k in range(courier):
            solver.add(couriers_loads[k] ==
                       Sum([If(x[k][i] == c, item_size[c-1], 0) for i in range(items+1) for c in range(1,items+1)]))

      #SImmetry breaking constraint
      for k in range(courier):
            solver.add(couriers_loads[k] <= courier_size[k])

      for k in range(courier):
        solver.add(x[k][0] != items+1)

      for i in range(courier-1):
        for j in range(1,items+1):
         solver.add(Implies(x[i][j]==items+1,x[i+1][j]==items+1))

      for k in range(courier):
            terms = [
                If(x[k][0] == c, distances[items][c-1], 0) for c in range(1,items+2)
            ] + [
                If(And(x[k][i] == d, x[k][i+1] == e), distances[d-1][e-1], 0)
                for d in range(1, items + 2) for e in range(1, items + 2) for i in range(items)
            ]

            solver.add(m_dist[k] == Sum(terms))

      # Constraint the maximum to be the maximum distance travelled among couriers
      dist = [m_dist[i] for i in range(courier)]
      max = dist[0]
      for d in dist[1:]:
            max = If(d > max, d, max)
      solver.add(maximum == max)

      print("Time needed to encode the instance:", t.time() - start_time)
      TIMEOUT = 300 * 1000
      solver.set('timeout', TIMEOUT)

      if solver.check() == unsat:
            return "Failed to solve"

      solver.push()

      UPPER_BOUND = up_bound
      LOWER_BOUND = low_bound

      TIMEOUT = 300 * 1000
      solver.set('timeout', TIMEOUT)

      if solver.check() == unsat:
        return "Failed to solve"

      satisfiable = True

      previousModel = None

      start_time = t.time()

      while(satisfiable):
        MIDDLE_BOUND =(int(str(UPPER_BOUND)) + LOWER_BOUND) // 2
        solver.add(maximum<=MIDDLE_BOUND)
        print(f"search inside [{LOWER_BOUND}-{MIDDLE_BOUND}]: ", end= "")

        current_time = t.time()
        past_time = int((current_time - start_time)*1000)
        solver.set('timeout', TIMEOUT - past_time)

        status = solver.check()

        print(status,end = "")
        if status == sat:
          model = solver.model()
          previousModel = model
          UPPER_BOUND = model.evaluate(maximum)
          print(UPPER_BOUND)

        elif status == unsat:
          solver.pop()
          solver.push()
          if LOWER_BOUND == MIDDLE_BOUND:
            MIDDLE_BOUND += 1
          LOWER_BOUND = MIDDLE_BOUND

        elif status == unknown:
          print(f"The computation time exceeded the limit ({TIMEOUT // 1000} seconds)")
          satisfiable = False

        if (int(str(UPPER_BOUND)) - int(str(LOWER_BOUND))) == 0:
          satisfiable = False


      current_time = t.time()
      past_time = int((current_time - start_time)*1000)

      print("Time from beginning of the computation: ", past_time//1000)
      model = previousModel

      current_time = t.time()
      past_time = int((current_time - start_time)*1000)
      if past_time > 300 * 1000:
            print(f"The computation time exceeded the limit ({TIMEOUT // 1000} seconds)")
      print("Time from beginning of the computation: ", past_time//1000)

      model = previousModel

      assignments = [[model.evaluate(x[k][j]) for j in range(items + 1)] for k in range(courier)]
      distances = np.array([model.evaluate(m_dist[i]) for i in range(courier)])
      loads = [model.evaluate(couriers_loads[k]) for k in range(courier)]
      final_maximum = model.evaluate(maximum).as_long()

      print(np.array(assignments))
      print(f"Travelled distances: {distances}\n")
      print(f"Final objective: {final_maximum}\n")
      print(f"Couriers load: {loads}")

mtps(instance)
```
